```sql
-- Criar a Tabela Silver para Produtos (Silver_Product)
    DROP TABLE IF EXISTS Silver_Product;
    CREATE TABLE Silver_Product AS
    SELECT 
        ProductID,
        Name AS ProductName,
        ProductNumber,
        Color,
        StandardCost,
        ListPrice,
        SellStartDate,
        SellEndDate
    FROM SalesLT.Product
    WHERE SellEndDate IS NULL; -- Apenas produtos ativos

    -- Criar a Tabela Silver para Clientes (Silver_Customer)
    DROP TABLE IF EXISTS Silver_Customer;
    CREATE TABLE Silver_Customer AS
    SELECT 
        CustomerID,
        CONCAT(FirstName, ' ', LastName) AS FullName,
        CompanyName,
        EmailAddress,
        Phone
    FROM SalesLT.Customer
    WHERE EmailAddress IS NOT NULL; -- Apenas clientes com email

    -- Criar a Tabela Silver para Pedidos (Silver_SalesOrderHeader)
    DROP TABLE IF EXISTS Silver_SalesOrderHeader;
    CREATE TABLE Silver_SalesOrderHeader AS
    SELECT 
        SalesOrderID,
        CustomerID,
        CAST(OrderDate AS DATE) AS OrderDate,
        CAST(ShipDate AS DATE) AS ShipDate,
        SubTotal,
        TaxAmt,
        Freight,
        TotalDue
    FROM SalesLT.SalesOrderHeader
    WHERE Status <> 4; -- Remove pedidos cancelados

    -- Criar a Tabela Silver para Detalhes dos Pedidos (Silver_SalesOrderDetail)
    DROP TABLE IF EXISTS Silver_SalesOrderDetail;
    CREATE TABLE Silver.SalesOrderDetail AS
    SELECT 
        SalesOrderID,
        ProductID,
        OrderQty,
        UnitPrice,
        UnitPriceDiscount,
        (UnitPrice * OrderQty) AS TotalLine
    FROM SalesLT.SalesOrderDetail
    WHERE OrderQty > 0; -- Remove quantidades inválidas

    PRINT 'Tabelas Silver criadas com sucesso!';

```
